**A priori algorthm**

```
F_k: frequent k-itemsets

L_k: candidate k-itemsets

Algorithm
* Let k=1
* Generate F_1 = {frequent 1-itemsets}
* Repeat until F_k is empty
 * Candidate Generation: Generate L_(k+1) from F_k
 * Candidate Pruning: eliminate candidate itemsets in L_k+1 containing subsets of length k that are infrequent 
 * Support Counting: Count the support of each candidate in L_(k+1) by scanning the DB (or use a more efficient hash method)
 * Candidate Elimination: Eliminate candidates in L_(k+1) that are infrequent, leaving only those that are frequent => F_(k+1)
```





````python
#Functions and imports
#I had functions for candidate generation, candidate pruning, support count, and candidate elimination.
from itertools import combinations

def candidate_generation(F_k, F_1):
  items = []
  item_list = set()
  for item_k in F_k:
    for item_1 in F_1:
      if item_1 not in item_k:
        items.append(item_1)
        if F_k == F_1:
          items.append(item_k)
        else:
          for product in item_k:
            items.append(product)
        #print(items)
        items.sort()
        item_list.add(tuple(items))
      items = []
  
  return item_list

def candidate_pruning(cand_list, freq_item, n):
  pruned_list = []
  #print("candidate list is {}".format(cand_list))
  #print("freq_item is {}".format(freq_item))
  for cand in cand_list:
    is_frequent = True
    sub_list = list(combinations(cand, n))
    
    for sub_can in sub_list:
      #print("sun can is {}".format(sub_can))
      if n == 1 and not set((sub_can)).issubset(set(freq_item)):
        is_frequent = False
        break
      elif n != 1:
        if sub_can not in freq_item:
          is_frequent = False
          break
    if is_frequent:
      pruned_list.append(cand)
  
  return pruned_list

def generate_frequent_itemset_n(freq_item, n):
  # candidate generation 
  item_list = candidate_generation(freq_item[n-1], freq_item[1])
  item_list = set(item_list)

  print("Candidate for {} is {}".format(n, item_list))
  output.write("Candidate for {} is {} \n".format(n, item_list))
  print("Len of candidate {}".format(len(item_list)))
  output.write("Len of candidate {} \n".format(len(item_list)))

  # candidate pruning
  item_pruned_list = candidate_pruning(item_list, freq_item[n-1], n-1)
  print("After prune {}".format(item_pruned_list))
  output.write("After prune {} \n".format(item_pruned_list))
  print("Len after prune {}".format(len(item_pruned_list)))
  output.write("Len after prune {} \n".format(len(item_pruned_list)))

  # support counting
  item_set = dict()
  for trans in data_matrix:
    for items in item_pruned_list:
      if(set(items).issubset(set(trans))):
        item_set[items] = item_set[items] + 1 if items in item_set else 1

  # candidate elimination
  freq_item_set = {k: v for k, v in item_set.items() if v >= MIN_SUP}
  return freq_item_set
````
